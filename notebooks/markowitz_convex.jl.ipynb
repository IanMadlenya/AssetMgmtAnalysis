### Portfolio weights

get portfolio moments for given weights: weights may have columns to label them, function only makes use of columns that have same labels as estimated moments

```julia
function getPfMoments(wgts::DataFrame, mus::DataFrame, covMatr::DataFrame)
    # get dimensions
    nPfs = size(wgts, 1)
    
    # get values without metadata
    muArr = convert(Array, mus)
    covMatrArr = convert(Array, covMatr)
    wgtsArr = convert(Array, wgts[:, names(mus)])
    
    # get columns with additional information
    infoColNams = setdiff(names(wgts), names(mus))
    
    # preallocate
    pfMus = zeros(nPfs)
    pfSigmas = zeros(nPfs)
    
    for ii=1:nPfs
        currWgts = wgtsArr[ii, :]
        currMu = currWgts*muArr'
        currSigma = sqrt(currWgts*covMatrArr*currWgts')
        pfMus[ii] = currMu[1]
        pfSigmas[ii] = currSigma[1]
    end
    pfMoments = DataFrame(mu = pfMus, sigma = pfSigmas)
    return [wgts[infoColNams] pfMoments]
end
```

get individual assets as portfolio weights with label **Asset**

```julia
function assetPortfolios(assNames::Array{Symbol, 1})
    # represent individual assets artificially as portfolios
    nAss = size(assNames, 1)
    wgtsArr = eye(nAss)
    wgts = composeDataFrame(wgtsArr, assNames)
    typs = String["Asset" for ii=1:nAss]
    descriptions = DataFrame(name = AssetMgmt.symbToStr(assNames), pfType = typs)
    return [descriptions wgts]
end
```

global minimum variance portfolio **without shortselling**

```julia
function noShortGmv(covMatr::DataFrame)
    # get covariance matrix without metadata
    covMatrArr = asArr(covMatr, Float64)
    
    # rescale to percentage return moments
    covMatrArr = 100^2 * covMatrArr
    
    # set up variable
    nAss = size(covMatrArr, 1)
    wgts = Variable(nAss)
    
    # define minimization problem
    addUpConstraint = sum(wgts) == 1
    noShortSellConstraint = wgts > 0
    noShortGmvOpt = minimize(quad_form(wgts, covMatrArr), addUpConstraint, noShortSellConstraint)
    
    # solve minimzation problem
    solve!(noShortGmvOpt, Convex.SCSSolver(verbose=false, rho_x = 1e-10))
    
    # get optimal weights
    xNoShortGmv = wgts.value
    
    return [DataFrame(name = "noShortGmv", pfType = "GMV") composeDataFrame(xNoShortGmv', names(covMatr))]
end
```

global minimum variance portfolio **with shortselling**

```julia
function gmv(covMatr)
    # get covariance matrix without metadata
    covMatrArr = asArr(covMatr, Float64)
    
    # rescale to percentage return moments
    covMatrArr = 100^2 * covMatrArr
    
    # set up variable
    nAss = size(covMatrArr, 1)
    wgts = Variable(nAss)
    
    # define minimization problem
    addUpConstraint = sum(wgts) == 1
    
    shortGmv = minimize(quad_form(wgts, covMatrArr), addUpConstraint)
    
    # solve minimzation problem
    solve!(shortGmv, Convex.SCSSolver(verbose=false, rho_x = 1e-10))
    
    # get optimal weights
    xShortGmv = wgts.value
    
    return [DataFrame(name = "shortGmv", pfType = "GMV") composeDataFrame(xShortGmv', names(covMatr))]
    
    
end
```

minimize variance with **given mu** without shortselling

```julia
function noShortMinSigma(mus::DataFrame, covMatr::DataFrame, givMu::Float64)
    # get covariance matrix without metadata
    muArr = asArr(mus, Float64)
    covMatrArr = asArr(covMatr, Float64)
    
    # rescale to percentage return moments
    muArr = 100*muArr
    covMatrArr = 100^2 * covMatrArr
    givMuScaled = 100*givMu
    
    nAss = size(covMatrArr, 1)
    wgts = Variable(nAss)
    
    # define minimization problem
    addUpConstraint = sum(wgts) == 1
    givenMuConstraint = wgts'*muArr' == givMuScaled
    noShortSellConstraint = wgts > 0
    minSigma = minimize(quad_form(wgts, covMatrArr), addUpConstraint, givenMuConstraint, noShortSellConstraint)
    
    # solve minimzation problem
    solve!(minSigma, Convex.SCSSolver(verbose=false, rho_x = 1e-10))
    
    # get optimal weights
    minSigmaWgts = wgts.value
    
    return [DataFrame(name = "mu_$givMu", pfType = "minSigma") composeDataFrame(minSigmaWgts', names(covMatr))]
end
```

get **efficient frontier** without shortselling

```julia
function noShortEff(mus::DataFrame, covMatr::DataFrame, nPfs::Int = 50)
    # calculate efficient frontier weights 
    
    # get no short selling GMV weights
    xNoShortGmv = noShortGmv(covMatr)
    
    # get associated portfolio mu
    xx = getPfMoments(xNoShortGmv, mus, covMatr)
    noShortGmvMu = xx[1, :mu]
    
    # get maximum asset mu
    maxMu = maximum(convert(Array, mus))
    
    # get linear grid for mus
    muGrid = linspace(noShortGmvMu, maxMu, nPfs)
    
    # preallocate
    nAss = size(covMatr, 1)
    wgtsNams = Array(String, nPfs)
    #wgtsTypes = Array(String, nPfs)
    wgtsTypes = String["noShortEff" for ii=1:nPfs]
    wgtsArr = zeros(nPfs, nAss)
    for ii=1:nPfs
        currGivenMu = muGrid[ii]
        currEffWgts = noShortMinSigma(mus, covMatr, currGivenMu)
        wgtsNams[ii] = currEffWgts[1, :name]
        #wgtsTypes[ii] = currEffWgts[1, :pfType]
        wgtsArr[ii, :] = convert(Array, currEffWgts[1, names(mus)])
    end
    
    return [DataFrame(name = wgtsNams, pfType = wgtsTypes) composeDataFrame(wgtsArr, names(mus))]
end
```

get **maximum sharpe ratio** portfolio: does not work that way, as it can not be proved to be a convex optimization problem this way

```julia
function noShortMaxSharpeFailing(mus::DataFrame, covMatr::DataFrame)
    # get covariance matrix without metadata
    muArr = asArr(mus, Float64)
    covMatrArr = asArr(covMatr, Float64)
    
    # rescale to percentage return moments
    muArr = 100*muArr
    covMatrArr = 100^2 * covMatrArr
    givMuScaled = 100*givMu
    
    nAss = size(covMatrArr, 1)
    wgts = Variable(nAss)
    
    # define minimization problem
    addUpConstraint = sum(wgts) == 1
    noShortSellConstraint = wgts > 0
    objFun = muArr*wgts / quad_form(wgts, covMatrArr)
    maxSharpe = maximize(objFun, addUpConstraint, noShortSellConstraint)
    
    # solve minimzation problem
    solve!(maxSharpe, Convex.SCSSolver(verbose=false, rho_x = 1e-10))
    
    # get optimal weights
    maxSharpeWgts = wgts.value
    
    return [DataFrame(name = "maxSharpe", pfType = "maxSharpe") composeDataFrame(maxSharpeWgts', names(covMatr))]
end
```

**maximum sharpe ratio** portfolio: using a trick to make optimization more easy

```julia
function noShortMaxSharpe(mus::DataFrame, covMatr::DataFrame)
    # get covariance matrix without metadata
    muArr = asArr(mus, Float64)
    covMatrArr = asArr(covMatr, Float64)
    
    # rescale to percentage return moments
    muArr = 100*muArr
    covMatrArr = 100^2 * covMatrArr
    givMuScaled = 100*givMu
    
    nAss = size(covMatrArr, 1)
    wgts = Variable(nAss)
    
    # define minimization problem
    muConstraint = muArr*wgts == 1
    noShortSellConstraint = wgts > 0
    maxSharpe = minimize(quad_form(wgts, covMatrArr), muConstraint, noShortSellConstraint)
    
    # solve minimzation problem
    solve!(maxSharpe, Convex.SCSSolver(verbose=false, rho_x = 1e-10))
    
    # rescale to sum up to 1
    maxSharpeWgts = wgts.value/sum(wgts.value)
    
    return [DataFrame(name = "maxSharpe", pfType = "maxSharpe") composeDataFrame(maxSharpeWgts', names(covMatr))]
end
```

function trying to determine **feasible set** of portfolios. However, the right fringe does not resemble a convex hull, such that optimization completely fails on right fringe of the frontier

```julia
function noShortFeasible(mus::DataFrame, covMatr::DataFrame, nPfs::Int = 50)
    # calculate feasible set weights
    
    # get covariance matrix without metadata
    muArr = asArr(mus, Float64)
    covMatrArr = asArr(covMatr, Float64)
    
    # get sigma range
    nAss = size(covMatr, 1)
    minSig = sqrt(minimum(Float64[covMatrArr[ii, ii] for ii=1:nAss]))
    maxSig = sqrt(maximum(Float64[covMatrArr[ii, ii] for ii=1:nAss]))
    
    # get linear grid for mus
    sigmaGrid = linspace(minSig, maxSig, nPfs)
    
    # preallocate
    wgtsNams = Array(String, nPfs)
    wgtsTypes = String["noShortFeasible" for ii=1:nPfs]
    wgtsArr = zeros(nPfs, nAss)
    for ii=1:nPfs
        currGivenSigma = sigmaGrid[ii]
        currWgts = noShortMaxMu(mus, covMatr, currGivenSigma)
        wgtsNams[ii] = currWgts[1, :name]
        #wgtsTypes[ii] = currEffWgts[1, :pfType]
        wgtsArr[ii, :] = convert(Array, currWgts[1, names(mus)])
    end
    
    return [DataFrame(name = wgtsNams, pfType = wgtsTypes) composeDataFrame(wgtsArr, names(mus))]
end
```

maximize mu with **given sigma**

```julia
function noShortMaxMu(mus::DataFrame, covMatr::DataFrame, givSigma::Float64)
    # get covariance matrix without metadata
    muArr = asArr(mus, Float64)
    covMatrArr = asArr(covMatr, Float64)
    
    # rescale to percentage return moments
    muArr = 100*muArr
    covMatrArr = 100^2 * covMatrArr
    givSigmaScaled = 100*givSigma
    givVar = givSigmaScaled^2
    
    nAss = size(covMatrArr, 1)
    wgts = Variable(nAss)
    
    # define minimization problem
    addUpConstraint = sum(wgts) == 1
    noShortSellConstraint = wgts > 0
    sigmaConstraint = quad_form(wgts, covMatrArr) == givVar
    maxMu = maximize(muArr*wgts, addUpConstraint, noShortSellConstraint, sigmaConstraint)
    
    # solve minimzation problem
    #solve!(maxMu, Convex.SCSSolver(verbose=false, rho_x = 1e-10, eps = 1e-10))
    solve!(maxMu, Convex.SCSSolver(verbose=false, normalize=0))
    
    # get optimal weights
    maxMuWgts = wgts.value
    
    return [DataFrame(name = "sigma_$givSigma", pfType = "maxMu") composeDataFrame(maxMuWgts', names(covMatr))]
end
```

same functionality, but using **NLopt** as optimizer. Any point near the right fringe seems to be hard to optimize and could also crucially depend on the starting point. A better implementation might use the closest asset as starting point.

```julia
using NLopt

function sigmaDeviation(wgts::Array{Float64, 1}, givSigma::Float64, covMatrArr::Array{Float64, 2})
    # get portfolio sigma
    pfSigma = sqrt(wgts'*covMatrArr*wgts)
    
    return (pfSigma[1] - givSigma)^2
end

function muDeviation(wgts::Array{Float64, 1}, givMu::Float64, muArr::Array{Float64, 1})
    # get portfolio mu
    pfMu = muArr*wgts
    
    return (pfMu[1] - givMu)^2
end

function momentDeviation(wgts::Array{Float64, 1}, givMu::Float64, 
    givSigma::Float64, muArr::Array{Float64, 1}, covMatrArr::Array{Float64, 1})
   
    return sigmaDeviation(wgts, givSigma, covMatrArr) + muDeviation(wgts, givMu, muArr)
end

function noShortMaxMuNLopt(mus::DataFrame, covMatr::DataFrame, givSigma::Float64)
    # get covariance matrix without metadata
    muArr = asArr(mus, Float64)
    covMatrArr = asArr(covMatr, Float64)
    
    # rescale to percentage return moments
    muArr = 100*muArr
    covMatrArr = 100^2 * covMatrArr
    givSigmaScaled = 100*givSigma
    
    nAss = size(covMatrArr, 1)
    
    opt = Opt(:LN_COBYLA, nAss)
    
    function objFun(x::Vector, grad::Vector)
        if length(grad) > 0
                ## no partial derivative given
        end
        return (muArr*x)[1]
    end
    
    max_objective!(opt, objFun)
    xtol_abs!(opt,1e-14)
    maxeval!(opt, 200000)

    ## constraints
    givenSigmaConstraint(x, g) = sigmaDeviation(x, givSigmaScaled, covMatrArr)
    equality_constraint!(opt, givenSigmaConstraint, 1e-13)

    addUpConstraint(x, g) = sum(x) - 1
    equality_constraint!(opt, addUpConstraint, 1e-13)

    ## short selling constraints
    lower_bounds!(opt, zeros(nAss))
    upper_bounds!(opt, ones(nAss))

    ## come up with initial values
    initX = 1/nAss*ones(nAss)

    (minf, minx, ret) = optimize(opt, initX)
    
    return [DataFrame(name = "sigma_$givSigma", pfType = "maxMuNLopt") composeDataFrame(minx', names(covMatr))]
end
```

### Match portfolio moments

try to match given portfolio moments with some portfolio: works nice for portfolios on left fringe, but right fringe is vastly failing

```julia
function noShortMinDist(mus::DataFrame, covMatr::DataFrame, muGiv::Float64, sigGiv::Float64)
    # match portfolio moments through Convex.jl
    
    # get covariance matrix without metadata
    muArr = asArr(mus, Float64)
    covMatrArr = asArr(covMatr, Float64)
    
    # rescale to percentage return moments
    muArr = 100*muArr
    covMatrArr = 100^2 * covMatrArr
    givMuScaled = 100*muGiv
    givSigmaScaled = 100*sigGiv
    scaleFactor = 1
    
    nAss = size(covMatrArr, 1)
    wgts = Variable(nAss)
    
    # define minimization problem
    addUpConstraint = sum(wgts) == 1
    noShortSellConstraint = wgts > 0
    objFun = (quad_form(wgts, covMatrArr) - givSigmaScaled^2)^2 + 
        scaleFactor*(muArr*wgts - givMuScaled)^2
    #objFun = (quad_form(wgts, covMatrArr) - givSigmaScaled^2)^2
    minDist = minimize(objFun, addUpConstraint, noShortSellConstraint)
    
    # solve minimzation problem
    solve!(minDist, Convex.SCSSolver(verbose=false, max_iters = 20000, rho_x = 1e-10, eps = 1e-10))
    
    # get weights
    minDistWgts = wgts.value
    
    return [DataFrame(name = "minDist", pfType = "minDist") composeDataFrame(minDistWgts', names(covMatr))]
end
```

same functionality, but three different ways to achieve matching portfolio:
- take specified mu as given, minimize deviation of sigma
- take specified sigma as given, minimize deviation of mu
- minimize deviation of sigma and mu

```julia
function noShortMinDistExt(mus::DataFrame, covMatr::DataFrame, muGiv::Float64, sigGiv::Float64)
    # extended version to replicate given moments
    
    # get covariance matrix without metadata
    muArr = asArr(mus, Float64)
    covMatrArr = asArr(covMatr, Float64)
    
    # rescale to percentage return moments
    muArr = 100*muArr
    covMatrArr = 100^2 * covMatrArr
    givMuScaled = 100*muGiv
    givSigmaScaled = 100*sigGiv
    givVar = givSigmaScaled^2
    
    nAss = size(covMatrArr, 1)
    wgts = Variable(nAss)
    
    # minimize distance given sigma
    addUpConstraint = sum(wgts) == 1
    noShortSellConstraint = wgts > 0
    sigmaConstraint = quad_form(wgts, covMatrArr) == givVar
    squDev = (muArr*wgts - givMuScaled)^2
    minMuDev = minimize(squDev, addUpConstraint, noShortSellConstraint, sigmaConstraint)
    
    # solve minimzation problem
    solve!(minMuDev, Convex.SCSSolver(verbose=false, max_iters = 20000, rho_x = 1e-8, eps = 1e-8))
    println("given sigma: $(minMuDev.status)")
    
    # 
    minMuDevWgts = wgts.value
    
    # reset wgts
    wgts = Variable(nAss)
    
    # minimize distance given mu
    addUpConstraint = sum(wgts) == 1
    givenMuConstraint = wgts'*muArr' == givMuScaled
    noShortSellConstraint = wgts > 0
    squDev = (quad_form(wgts, covMatrArr) - givVar)^2
    minSigmaDev = minimize(squDev, addUpConstraint, noShortSellConstraint, givenMuConstraint)
    
    # solve minimization problem
    solve!(minSigmaDev, Convex.SCSSolver(verbose=false, max_iters = 20000, rho_x = 1e-8, eps = 1e-8))
    println("given mu: $(minSigmaDev.status)")
    
    # get weights
    minSigmaDevWgts = wgts.value
    
    # min both
    minBoth = noShortMinDist(mus, covMatr, muGiv, sigGiv)
    minBothWgts = minBoth[1, names(mus)]
    minBothWgts = convert(Array, minBothWgts)
    
    # collect weights
    allWgts = [minMuDevWgts'; minSigmaDevWgts'; minBothWgts]
    
    nams = ["minMuDev"; "minSigmaDev"; "minBothDev"]
    
    return [DataFrame(name = nams, pfType = "minDist") composeDataFrame(allWgts, names(covMatr))]
end
```

#### Efficient frontier plot

```julia
function getUniverseInfo(mus::DataFrame, covMatr::DataFrame; nPts::Int = 50)
    
    
    # get weights for assets
    assWgts = assetPortfolios(names(mus))
    
    # get global minimum variance portfolios
    xNoShortGmv = noShortGmv(covMatr)
    xShortGmv = gmv(covMatr)
    
    # get efficient frontier weights and max sharpe ratio weights
    xNoShortEff = noShortEff(mus, covMatr, nPts)
    maxSharpeWgts = noShortMaxSharpe(mus, covMatr)
    
    # collect all weights
    allWgts = [assWgts; xNoShortGmv; xShortGmv; xNoShortEff; maxSharpeWgts]
    
    # get moments for each portfolio
    pfMoments = getPfMoments(allWgts, mus, covMatr)
    
    # get scaled annualized moments for each portfolio
    pfMus = convert(Array, pfMoments[:mu])
    pfSigmas = convert(Array, pfMoments[:sigma])
    scaledMus, scaledSigmas = AssetMgmt.defaultMuSigmaScaling(pfMus, pfSigmas)
    
    # attach to portfolio moments
    pfMoments[:annual_mus] = scaledMus
    pfMoments[:annual_sigmas] = scaledSigmas
    
    # join weights
    universeVals = join(pfMoments, allWgts, on = [:name, :pfType])
    
    return universeVals
end
```

```julia
function plotEfficientFrontier(mus::DataFrame, covMatr::DataFrame; nPts::Int = 50)
    # visualize current universe by depicting efficient frontier
    
    # get relevant portfolios
    universeInfo = getUniverseInfo(mus, covMatr, nPts = nPts)
    
    # split into efficient frontier and other portfolios
    xxEff = universeInfo[universeInfo[:pfType] .== "noShortEff", :]
    xxOther = universeInfo[universeInfo[:pfType] .!= "noShortEff", :]

    # create plot
    p = plot(layer(xxEff, x=:annual_sigmas, y=:annual_mus, Geom.line), 
        layer(xxOther, x=:annual_sigmas, y=:annual_mus, color=:pfType, label=:name, Geom.point, Geom.label), 
        Guide.XLabel("sigma"), Guide.YLabel("mu"))
    
    return p
end
```

set up function to transform collected sigmas and correlations back to covariance matrix

```julia
function covMatrCollectionToMatrix(sigmas::Array{Float64, 2}, corrs::Array{Float64, 2})
# get indices of lower and upper matrix triangles 
    nAss = size(sigmas, 2)
nEntries = int(nAss*(nAss-1)/2)
indLower = zeros(Int, nEntries)
indUpper = zeros(Int, nEntries)
    matrDims = (nAss, nAss)
entryCounter = 1
for ii=1:(nAss-1)
    for jj=ii+1:nAss
        indLower[entryCounter] = sub2ind(matrDims, jj, ii)
        indUpper[entryCounter] = sub2ind(matrDims, ii, jj)
        entryCounter += 1
    end
end
    
    # fill matrix
newCorrMatr = zeros(nAss, nAss)
for ii=1:size(xxCorrs, 1)
    lInd = indLower[ii]
    uInd = indUpper[ii]
    newCorrMatr[lInd] = xxCorrs[ii]
    newCorrMatr[uInd] = xxCorrs[ii]
end
    
    # translate correlations to covariances
d = diagm(sigmas[:])
newCovMatr = d*newCorrMatr*d
    
    # fill in diagonal elements
for ii=1:nAss
    newCovMatr[ii, ii] = sigmas[ii]^2
end
    return newCovMatr 
end
```

set up function to check `covMatrCollectionToMatrix`

```julia
function collectCovMatrEntries(covMatr::DataFrame)

covMatrArr = convert(Array, covMatr)

nAss = size(covMatrArr, 1)
sigmas = (Float64[sqrt(covMatrArr[jj, jj]) for jj=1:nAss])'

## get correlation matrix
d = diagm(1./sqrt(diag(covMatrArr)))
corrMatr = d*covMatrArr*d

## extract correlations
corrs = vcat([corrMatr[(jj+1:end), jj] for jj=1:(nAss-1)]...)

return sigmas, corrs

end
```
